# Multi plot grids

```python
c = sns.load_dataset('iris')

# WIth facetgrid i specify the parameter i wanna differentate with
d = sns.FacetGrid(c,col='species')
# Here, i specify the kind of plot i want for all and the param that is to be varied
d.map(plt.hist,"sepal_length")
```

```python
e = sns.load_dataset('flights')
f = sns.PairGrid(e)
f.map(plt.scatter)
```

```python
sns.set(style="darkgrid",color_codes=True)
e = sns.load_dataset('flights')
f = sns.PairGrid(e)
f.map(plt.scatter)
```
